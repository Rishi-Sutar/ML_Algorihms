```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Experience-Salary.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(X,y,test_size =0.2)
```

```python
# Plot the data points
plt.scatter(x_train,y_train, marker='x', c='r')
# Set the title
plt.title("Salary vs. Experience")
# Set the y-axis label
plt.ylabel('Salary')
# Set the x-axis label
plt.xlabel('Experience')
plt.show()
```

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_predict=regressor.predict(x_test)
y_predict
```

```python
y_test
```

```python
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train,regressor.predict(x_train), color='blue')
plt.title("Linear Regression salary vs experinces", size=18)
plt.xlabel("Year of Employee",size=15)
plt.ylabel("Saleries of Employee", size=15)
plt.show()
```

```python
# now lets predict for test

plt.scatter(x_test, y_test, color='red')
plt.plot(x_test,regressor.predict(x_test),color='Blue')
plt.title("linear regression salary vs experince", size=16)
plt.xlabel("Year of Employee", size=12)
plt.ylabel("Saleries of Employee", size=12)
plt.show()
```
